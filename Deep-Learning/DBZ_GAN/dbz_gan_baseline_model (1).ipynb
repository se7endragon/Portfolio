```python
#Data Import from gdrive and resize the image
train_img = []
train_labels = []
path = 'dataset'

for filename in os.listdir(path):
  if filename.startswith("s"):
    if filename.endswith("jpg") or filename.endswith("JPG") or filename.endswith("png") or filename.endswith("jpeg") or filename.endswith("PNG"):
      resized = tf.image.resize(cv2.imread(os.path.join(path,filename)),(64,64),antialias=True)
      train_img.append(resized)
      if filename[3] == 'b':
        train_labels.append([4])
      else:
        train_labels.append([int(filename[3])-1])

dataset = np.ndarray(shape=(len(train_img),64,64,3),dtype=np.float32)

i = 0
for image in train_img:
  x = tf.keras.preprocessing.image.img_to_array(image)
  x = x.reshape((64,64,3))
  dataset[i] = x
  i += 1

train_labels = np.array(train_labels)
```

```python
BUFFER_SIZE = len(dataset)
BATCH_SIZE = 64
```

```python
# Adding random noise on the original images
#dataset = tf.image.random_brightness(dataset, max_delta=0.1)
#dataset = tf.image.random_contrast(dataset, lower=0.9, upper=1.1)
# Normalize the images to [-1, 1]
train_images = (dataset - 127.5) / 127.5 

#Shuffle and make data into batches
train_dataset = tf.data.Dataset.from_tensor_slices((train_images,train_labels)).shuffle(BUFFER_SIZE).batch(BATCH_SIZE)
train_dataset
```

```python

def make_generator_model():
    label_in = tf.keras.Input(shape=(1,))
    emb = layers.Embedding(5, 25)(label_in)
    li = layers.Dense(4*4)(emb)
    li = layers.Reshape((4,4,1))(li)

    latent = tf.keras.Input(shape=(100,))
    gen = layers.Dense(4*4*512)(latent)
    gen = layers.Reshape((4,4,512))(gen)

    merged = layers.Concatenate()([gen,li])

    #Upsampling to 4x4x512
    model = layers.Conv2DTranspose(512, (4, 4), strides = (1,1), padding = 'same', use_bias=False)(merged)
    model = layers.BatchNormalization()(model)
    model = layers.ReLU()(model)

    #Upsampling to 8x8x256
    model = layers.Conv2DTranspose(256, (4, 4), strides = (2,2), padding = 'same', use_bias=False)(model)
    model = layers.BatchNormalization()(model)
    model = layers.ReLU()(model)

    #Upsampling to 16x16x128
    model = layers.Conv2DTranspose(128, (4, 4), strides = (2,2), padding = 'same', use_bias=False)(model)
    model = layers.BatchNormalization()(model)
    model = layers.ReLU()(model)
    
    #Upsampling to 32x32x64
    model = layers.Conv2DTranspose(64, (4, 4), strides = (2,2), padding = 'same', use_bias=False)(model)
    model = layers.BatchNormalization()(model)
    model = layers.ReLU()(model)

    #Upsampling to 64x64x3
    model = layers.Conv2DTranspose(3, (4, 4),strides = (2,2),padding = 'same', activation='tanh',use_bias=False)(model)
    model = tf.keras.Model([latent,label_in],model)
    return model
```

```python
generator = make_generator_model()
plot_model(generator,show_shapes=True)
```

```python
def make_discriminator_model():
    label_in = tf.keras.Input(shape=(1,))
    emb = layers.Embedding(5,25)(label_in)
    li = layers.Dense(64*64)(emb)
    li = layers.Reshape((64,64,1))(li)

    image = tf.keras.Input(shape=(64,64,3))
    
    merged = layers.Concatenate()([image,li])

    #Downsampling to 32x32x64
    x = layers.Conv2D(64, (4, 4),strides = (2,2),padding='same',use_bias=False)(merged)
    act = layers.LeakyReLU()(x)
    
    #downsampling to 16x16x128
    x = layers.Conv2D(128, (4, 4), strides=(2, 2), padding='same',use_bias=False)(act)
    x = layers.BatchNormalization()(x)
    act = layers.LeakyReLU()(x)
    
    #downsampling to 8x8x256
    x = layers.Conv2D(256, (4, 4), strides=(2, 2), padding='same',use_bias=False)(act)
    x = layers.BatchNormalization()(x)
    act = layers.LeakyReLU()(x)

    #downsampling to 4x4x512
    x = layers.Conv2D(512, (4, 4), strides=(2, 2), padding='same',use_bias=False)(act)
    x = layers.BatchNormalization()(x)
    act = layers.LeakyReLU()(x)
    output = layers.Flatten()(act)
    source_out = layers.Dense(1)(output)
    model = tf.keras.Model(inputs = [image,label_in], outputs=[source_out])
    return model
```

```python
discriminator = make_discriminator_model()
plot_model(discriminator,show_shapes=True)
```

```python
cross_entropy = tf.keras.losses.BinaryCrossentropy(from_logits=True)
```

```python
def discriminator_loss(real_output, fake_output):
    real_loss = cross_entropy(tf.ones_like(real_output), real_output)
    fake_loss = cross_entropy(tf.zeros_like(fake_output), fake_output)
    total_loss = real_loss + fake_loss
    #WGAN
    #total_loss = tf.reduce_mean(fake_output) - tf.reduce_mean(real_output)
    return total_loss

def generator_loss(fake_output):
    return cross_entropy(tf.ones_like(fake_output), fake_output)
    #WGAN
    #return -tf.reduce_mean(fake_output)
```

```python
generator_optimizer = tf.keras.optimizers.Adam(lr=0.0002, beta_1=0.5)
discriminator_optimizer = tf.keras.optimizers.Adam(lr=0.0002, beta_1=0.5)
#WGAN
#generator_optimizer = tf.keras.optimizers.RMSprop(lr=2e-4)
#discriminator_optimizer = tf.keras.optimizers.RMSprop(lr=2e-4)
```

```python
checkpoint_dir = './training_checkpoints'
checkpoint_prefix = os.path.join(checkpoint_dir, "ckpt")
checkpoint = tf.train.Checkpoint(generator_optimizer=generator_optimizer,
                                 discriminator_optimizer=discriminator_optimizer,
                                 generator=generator,
                                 discriminator=discriminator)
```

```python
EPOCHS = 4000
noise_dim = 100
num_examples_to_generate = 16
seed = tf.random.normal([num_examples_to_generate, noise_dim])
```

```python
def plot_history(d_hist, g_hist):
  plt.subplot(211)
  plt.plot(d_hist, label='disc')
  plt.subplot(212)
  plt.plot(g_hist, label='gen')
  plt.savefig('plot_line_plot_loss2.png')
  plt.close()
```

```python
@tf.function
def train_step(images, labels):
    noise = tf.random.normal([labels.shape[0], noise_dim])
    fake_labels = np.expand_dims(np.random.randint(0,5,labels.shape[0]),axis=1)
    with tf.GradientTape() as gen_tape, tf.GradientTape() as disc_tape:
      generated_images = generator([noise,fake_labels], training=True)
      real_output = discriminator([images,labels], training=True)
      fake_output = discriminator([generated_images,fake_labels], training=True)
      gen_loss = generator_loss(fake_output)
      disc_loss = discriminator_loss(real_output, fake_output)
    gradients_of_generator = gen_tape.gradient(gen_loss, generator.trainable_variables)
    gradients_of_discriminator = disc_tape.gradient(disc_loss, discriminator.trainable_variables)
    #capped_gradients = [(tf.clip_by_value(grad, -0.01, 0.01)) for grad in gradients_of_discriminator]
    generator_optimizer.apply_gradients(zip(gradients_of_generator, generator.trainable_variables))
    discriminator_optimizer.apply_gradients(zip(gradients_of_discriminator, discriminator.trainable_variables))
    return gen_loss, disc_loss
```

```python
def train(dataset, epochs, conditionals):
  d_loss, g_loss = list(), list()
  for epoch in range(epochs):
    start = time.time()
    for image_batch, label_batch in dataset:
      ge_loss, di_loss = train_step(image_batch, label_batch)
      g_loss.append(ge_loss)
      d_loss.append(di_loss)
    if (epoch +1) % 20 == 0:
        display.clear_output(wait=True)
        generate_and_save_images(generator, epoch + 1,seed, conditionals)
        print ('Time for epoch {} is {} sec'.format(epoch + 1, time.time()-start))
    # Save the model every 100 epochs
    if (epoch + 1) % 100 == 0:
        checkpoint.save(file_prefix = checkpoint_prefix)

  # Generate after the final epoch
  display.clear_output(wait=True)
  generate_and_save_images(generator,
                           epochs,
                           seed, conditionals)
  
  # plot the loss curve
  plot_history(d_loss, g_loss)
```

```python
def generate_and_save_images(model, epoch, test_input,conditionals):
  test_labels = np.expand_dims(np.random.randint(0,5,test_input.shape[0]),axis=1)
  predictions = model([test_input,test_labels], training=False)
  
  #[-1,1 to 0,1 for plotting]
  predictions = (predictions + 1)/2.0
  conditionals.append(test_labels)
  fig = plt.figure(figsize=(12,12))

  for i in range(predictions.shape[0]):
      plt.subplot(4, 4, i+1)
      plt.imshow(predictions[i])
      plt.axis('off')

  plt.savefig('image_at_epoch_{:04d}.png'.format(epoch))
  plt.show()
```

```python
conditionals = []
train(train_dataset, EPOCHS, conditionals)
#The randomly assigned condition images will be saved to the local directory
```
